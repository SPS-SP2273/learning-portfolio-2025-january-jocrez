## Exercise 3

```python
squares1 = [i**2 for i in range(10)]
```

```python
print(squares1)
```

```python
squares2 = list()
for i in range(10):
    squares2.append(i**2)
```

```python
print(squares2)
```

```python
%%timeit
squares1
squares2
```

it takes on average 70 nanoseconds to execute both lines of code with a standard deviation of 8.81 nanoseconds. Each code was executed 10,000,000 times. The average time was taken with the best 7 runs
