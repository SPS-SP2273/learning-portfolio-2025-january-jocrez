```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
```

```python

fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,3), sharex = 'col', sharey = 'row')

ax[0,0].plot(x, cos_x, label='cos x')
ax[0,0].grid(alpha = 0.25)
ax[0,1].plot(x, cos_x, label='cos x')
ax[0,1].grid(alpha = 0.25)
ax[1,0].plot(x, sin_x, label='sin x', c='red')
ax[1,0].grid(alpha = 0.25)
ax[1,1].plot(x, sin_x, label='sin x', c='red')
ax[1,1].grid(alpha = 0.25)

ax[0,0].set_xlabel('heheh')

for axis in ax.flatten():
    axis.legend()

plt.tight_layout()
```
