# Task 2

```python
def rename_my_file(old_file_name):
    after = old_file_name.replace('_','-')
    letters = after[0:3]
    if letters == convert.keys():
        num = convert[letters]
        after = after.replace(letters, num)
        return after
    else:
        return 'File already renamed' 
```

```python
rename_my_file('jan-08_singapore.txt')
```

```python
for i in range(len(file_list)):
    old_file_name = file_list[i]
    new_file_name = rename_my_file(old_file_name)
    print(old_file_name)
```

# Task 4

```python
folder_path = r'os-collaboration-exercise-data_1'
txt_files = glob.glob(r'os-collaboration-exercise-data_1/*.txt')

for file_path in txt_files:
    try:
        old_path = file_path
        old_file_name = os.path.split(file_path)[-1]
        new_file_name = rename_my_file(old_file_name)
        if new_file_name == 'File already renamed':
            print('File already renamed')
        else:
            new_path = os.path.join(folder_path, new_file_name)
            os.rename(old_path, new_path)
            print(f'File renamed from {old_file_name} to {new_file_name}')
    except OSError:
        print("Uh Oh")
```

# Task 7

```python
def moving(Cities):
    city_files = glob.glob(f'os-collaboration-exercise-data_1\\*{Cities}.*')
    #print(f'os-collaboration-exercise-data_1\\*{Cities}.*')
    for source_path in city_files:
        file_name = source_path.split(os.path.sep)[-1]
       # print(file_name)
        destination_path = f'os-collaboration-exercise-data_1\\{Cities}\\{file_name}'
        #print(destination_path)
        if os.path.exists(destination_path):
            print('Path already exists')
        else:
            shutil.move(source_path,destination_path)
            print(f'Moved {source_path} into {destination_path}')
```

```python
for city in city_list:
    moving(city)
```
